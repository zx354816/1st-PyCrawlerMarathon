```python
import urllib.request
import zipfile

res = "http://opendata.cwb.gov.tw/govdownload?dataid=F-D0047-093&authorizationkey=rdec-key-123-45678-011121314"

urllib.request.urlretrieve(res,'./Data/example.zip')

f = zipfile.ZipFile('./Data/example.zip')
f.extractall('./Data')
```

```python
import xml.dom.minidom

doc = xml.dom.minidom.parse("./Data/64_72hr_CH.xml")

print(doc.getElementsByTagName("datasetDescription")[0].firstChild.nodeValue)

dataTime = doc.getElementsByTagName("dataTime")

for d in dataTime:
    print( d.getAttribute("name") , d.firstChild.nodeValue)
```
